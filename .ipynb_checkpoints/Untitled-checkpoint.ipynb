```python
def metrics(y_train, train_preds, y_test, test_preds, log=False):
    print('Train Scores:')
    print(f'R2: {r2_score(y_train, train_preds)}')
    if log == False:
        print(f'RMSE: {mean_squared_error(y_train, train_preds, squared=False)}')
    else:
        print(f'RMSE: {mean_squared_error(np.expm1(y_train), np.expm1(train_preds), squared=False)}')
    print('Test Scores:')
    print(f'R2: {r2_score(y_test, test_preds)}')
    if log == False:
        print(f'RMSE: {mean_squared_error(y_test, test_preds, squared=False)}')
    else:
        print(f'RMSE: {mean_squared_error(np.expm1(y_test), np.expm1(test_preds), squared=False)}')
```

```python
cat_cols = ['date', 'bedrooms', 'bathrooms', 'sqft_living', 'view', 'condition',
       'grade', 'sqft_above', 'yr_built', 'lat', 'long', 'sqft_living15',
       'waterfront', 'zipcode']
```
